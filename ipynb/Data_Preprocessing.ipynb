```python
#get the dataset and clean .

#get the column names to be in proper format, 
#standardize the column names
#extract year and month column as separate from archive_date

#2019 data
dataset = pd.read_csv('OP Waiting List By Group Hospital 2019.csv')
dataset.columns = dataset.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
dataset['year'] = pd.DatetimeIndex(dataset['archive_date']).year
dataset['month'] = pd.DatetimeIndex(dataset['archive_date']).month
print(dataset.head)


#2018 data
dataset18 = pd.read_csv('OP Waiting List By Group Hospital 2018.csv')
dataset18.columns = dataset18.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
print(dataset18.head)
dataset18['year'] = pd.DatetimeIndex(dataset18['archive_date']).year
dataset18['month'] = pd.DatetimeIndex(dataset18['archive_date']).month


#2017 data
dataset17 = pd.read_csv('OP Waiting List By Group Hospital 2017.csv')
dataset17.columns = dataset17.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
print(dataset17.head)
dataset17['year'] = pd.DatetimeIndex(dataset17['archive_date']).year
dataset17['month'] = pd.DatetimeIndex(dataset17['archive_date']).month
print(dataset17.head)

#2016 data
dataset16 = pd.read_csv('OP Waiting List By Group Hospital 2016.csv')
dataset16.columns = dataset16.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
print(dataset16.head)
dataset16['year'] = pd.DatetimeIndex(dataset16['archive_date']).year
dataset16['month'] = pd.DatetimeIndex(dataset16['archive_date']).month
print(dataset16.head)

```

```python
hospitallist= pd.read_csv('listofhospitalsinireland.csv')
```

```python
hospitallist.head()
```

```python
hospitallist['add'] = hospitallist[['address', 'country']].apply(lambda x: ' '.join(x), axis=1)
```

```python
hospitallist.head()
```

```python
import geopandas as gpd
import geopy
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

locator = Nominatim(user_agent="myGeocoder")

# 1 - conveneint function to delay between geocoding calls
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
# 2- - create location column
hospitallist['location'] = hospitallist['add'].apply(geocode)
# 3 - create longitude, laatitude and altitude from location column (returns tuple)
hospitallist['point'] = hospitallist['location'].apply(lambda loc: tuple(loc.point) if loc else None)
# 4 - split point column into latitude, longitude and altitude columns
hospitallist[['latitude', 'longitude', 'altitude']] = pd.DataFrame(hospitallist['point'].tolist(), index=hospitallist.index)
```

```python
hospitallist.head()
```

```python
#extract the final dataset to excel 
hospitallist.to_excel (r'HospitalList-final.xlsx', index = None, header=True)
```

```python
import geopandas as gpd
import geopy
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

locator = Nominatim(user_agent="myGeocoder", timeout =3)
# 1 - conveneint function to delay between geocoding calls
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
# 2- - create location column
phar['location'] = phar['Address'].apply(geocode)
# 3 - create longitude, laatitude and altitude from location column (returns tuple)
phar['point'] = phar['location'].apply(lambda loc: tuple(loc.point) if loc else None)
# 4 - split point column into latitude, longitude and altitude columns
phar[['latitude', 'longitude', 'altitude']] = pd.DataFrame(phar['point'].tolist(), index=phar.index)
```

```python
phar.head()
```

```python
from pyproj import Proj, transform

inProj = Proj(init='epsg:3857')
outProj = Proj(init='epsg:4326')
def convert(df):
    return transform(inProj,outProj,df['ig_xcord'],df['ig_ycord'])
```

```python

#phar['newlat'] = phar['ig_xcord'].apply(lambda x: ) 
phar[['newlat','newlong']] = pd.DataFrame(phar.apply(convert, axis=1),index=phar.index)
```

```python
#extract the final dataset to excel 
phar.to_excel (r'Pharmacies-final.xlsx', index = None, header=True)
```
